```python
df['City'] = df['Purchase Address'].apply(lambda x : x.split(',')[1])
df.head()
```

### columns to the correct type

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
df.head()
```

### ADDING A SALES COLUMN FOR THE ANALYSIS

```python
df['sales'] = df['Quantity Ordered'] * df['Price Each']
df.head()
```

### What Month was the best month for sales? and how much it was earned

```python
results = df.groupby('Months').sum()
```

```python
import matplotlib.pyplot as plt
Months = range (1,3)
plt.bar(Months, results['sales'])
plt.xticks(Months)
plt.ylabel('sales')
plt.xlabel('months')
plt.show()
```

### To check the duplicated data in the dataset and which products are sold together?


```python
df1 = df[df['Order ID'].duplicated(keep=False)]
df1.head()
```

```python
df1['grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df1 = df1[['Order ID', 'grouped']].drop_duplicates()
df1.head()
```

### Counting Pairs of Products Ordered 

```python
from itertools import combinations
from collections import Counter


Count = Counter()

for row in df1['grouped']:
    row_list = row.split(',')
    Count.update(Counter(combinations(row_list, 2)))

for key, value in Count.most_common(10):
    print(key, value)

```
